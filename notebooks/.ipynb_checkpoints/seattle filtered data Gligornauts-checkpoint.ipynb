```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../data/kc_house_data.csv')
```

```python
df['city'] = df['address'].str.split(',').str[1].str.strip()
```

```python
df['address'].head()[0]
```

```python
df['zipcode'] = df['address'].str.split(',').str[-2].str[-5:]
```

```python
df['zipcode'].head()
```

```python
df_irs = pd.read_csv('../data/income_by_zip.csv')
```

```python
df_wa = df_irs.loc[df_irs['STATE'] == 'WA',['STATE','ZIPCODE','N02650','A02650']]
```

```python
df_wa['avg_inc'] = df_wa['A02650']/df_wa['N02650']
```

```python
df['zipcode'] = df['zipcode'].astype(int)
```

```python
df2 = df.merge(df_wa,left_on='zipcode',right_on='ZIPCODE',how='inner')
```

```python
df2['month'] = pd.DatetimeIndex(df2['date']).month
```

```python
df2['year'] = pd.DatetimeIndex(df2['date']).year
```

```python
df2['seattle'] = np.where(df2['city'] == 'Seattle',1,0)
```

```python
df_rich = df2.loc[(1 < df2['bedrooms']) & (df2['bedrooms'] < 4) & (df2['seattle'] == 1) & (df2['sqft_lot'] < 10000)
                 & ((df2['grade'] != '4 Low') & (df2['grade'] != '12 Luxury') & (df2['grade'] != '11 Excellent')
                 & (df2['grade'] != '10 Very Good') & (df2['grade'] != '9 Better'))].copy()
```

```python
def inc_bins(row):
    if row['avg_inc'] < 100:
        return 0
    elif 100 <= row['avg_inc'] < 150:
        return 1
    else:
        return 2
```

```python
df_rich['avg_inc_bins'] = df_rich.apply(inc_bins,axis=1)
```

```python
df_rich['avg_inc_bins'].value_counts()
```

```python
df_rich_remove = df_rich.loc[(df_rich['price'] < 2000000) & (df_rich['avg_inc'] < 200000)]
```

```python
df_final = pd.get_dummies(data = df_rich_remove,columns = ['bedrooms','cond_group','grade','year','avg_inc_bins'],drop_first = True)
```

```python
df_final.head()
```

```python
df_final.corr()['price'].sort_values(ascending=False)
```

```python
y1 = df_final['price']
x1 = df_final['sqft_living']
model = sm.OLS(y1,sm.add_constant(x1)).fit()
model.summary()
```

```python
from statsmodels.tools.eval_measures import rmse
ypred = model.predict(sm.add_constant(x1))
rmse = rmse(y1, ypred)
rmse
```

```python
y_final = df_final['price']
x_final = pd.concat([df_final['sqft_living'],df_final.iloc[:,-8:-6],df_final.iloc[:,-3:]],axis=1)
model = sm.OLS(y_final,sm.add_constant(x_final)).fit()
model.summary()
```

```python
from statsmodels.tools.eval_measures import rmse
ypred = model.predict(sm.add_constant(x_final))
rmse = rmse(y_final, ypred)
rmse
```
