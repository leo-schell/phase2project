## How to Provide Immediate Relief for Seattle's Housing Affordability Crisis

## Business Understanding

Seattle's problem is it has a major housing shortage. There's specifically a shortage of affordable housing for low-income families. A big reason for that is that Seattle's middle-income families have been priced out of homeownership and down-rent affordable units. Frustrations over the housing crisis have reached a boiling point, and on February 14, Seattle voted Yes on Initiative 135, creating a new Seattle Social Housing Authority. The SSHA's main function is to provide housing--through construction and acquistion--for those making under the city's median income. Our proposal to the SSHA is to buy and rent out homes to middle-income families.

We are defining middle-income families as making 80-120\% of Seattle median income of \\$120k. Using a 30\% rent to income ratio means they can spend between \\$2400 and \\$3600 a month on rent. Median single family monthly rent is around \\$3000. Thus, if we can buy up median-priced homes, we would only need to discount up to \\$600 for the lower end middle-income families. This is much more affordable than if they were to try and purchase these homes themselves, as the median mortgage payment is over \\$5000 per month.


## Data Understanding

We are looking to estimate the budget needed for the SSHA to buy 1000 homes for middle-income families. To do this, we created a linear regression model that predicts price based on select characteristics of a home. We are using mid 2021 to mid 2022 King County home sales data in the city of Seattle for 2-3 bedroom homes. We also gathered tax return data from the IRS to create a proxy for average household income by zipcode. We filtered for homes under \\$2 million, located in zipcodes with mean incomes under \\$200k, and lot square footage under 10,000 to minimize budget costs.

```python
#imports appropriate packages
import statsmodels.api as sm
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reads in the King County home sales data .csv file
df = pd.read_csv('../data/kc_house_data.csv')
```

```python
#creates a new column of zipcodes
df['zipcode'] = df['address'].str.split(',').str[-2].str[-5:]
```

```python
#converts King County zipcode column to type integer to prepare for data merging
df['zipcode'] = df['zipcode'].astype(int)
```

```python
#determined average income should be binned
#creates a function that bins average income into 3 categories
def inc_bins(row):
    if row['avg_inc'] < 100:
        return 0
    elif 100 <= row['avg_inc'] < 150:
        return 1
    else:
        return 2
```

```python
#applies the function to create a new column of average zipcode income bins
df_rich['avg_inc_bins'] = df_rich.apply(inc_bins,axis=1)
```

## Modeling

First, we ran a baseline dummy regressor model using mean price as the explanatory variable. Unsurprisingly, this model wasn't great, with an R-squared of 0 and a Root Mean Squared Error (RMSE) of almost \\$300,000.

```python
#creates dummy regressor model using mean price
y0 = df_final['price']
x0 = df_final['mean_price']
model = sm.OLS(y0,sm.add_constant(x0)).fit()
model.summary()
```

```python
#generates RMSE
from statsmodels.tools.eval_measures import rmse
ypred = model.predict(sm.add_constant(x0))
rmse = rmse(y0, ypred)
rmse
```

### Simple Linear Regression Model

Next, we ran a simple linear regression model using living space square footage as the only explanatory variable. Living space square footage was chosen because it had the highest correlation with price. Our model has an R-squared of .299, meaning 29.9% of the variability in price can be explained by changes in living space square footage. The interpretation of the beta coefficient is as follows: a 1 square foot increase in living space is associated with on average a \\$351 increase in price. The RMSE for this model was about \\$250,000, meaning that on average our predicted prices would be off by \\$250,000 relative to the actual prices. The size of this error is not ideal, especially given that median home prices are around \\$750,000.

```python
#creates 1-variable linear regression model using living space square footage
y1 = df_final['price']
x1 = df_final['sqft_living']
model1 = sm.OLS(y1,sm.add_constant(x1)).fit()
model1.summary()
```

```python
#generates RMSE for simple linear regression
from statsmodels.tools.eval_measures import rmse
ypred1 = model1.predict(sm.add_constant(x1))
rmse1 = rmse(y1, ypred1)
rmse1
```

Linearity assumption appears to be met for this model, though it's not the strongest.

```python
#checks linearity assumption
preds1 = model1.predict(sm.add_constant(x1))
fig, ax = plt.subplots()

perfect_line = np.arange(y1.min(), y1.max())
ax.plot(perfect_line, linestyle="--", color="orange", label="Perfect Fit")
ax.scatter(y1, preds1, alpha=0.5)
ax.set_xlabel("Actual Price")
ax.set_ylabel("Predicted Price")
ax.legend();
```

This model appear to have violated the normality assumption. Specifically, we appear to have problems at the tails of the dataset. We may want to restrict the dataset further by excluding outliers at both extremes.

```python
#checks normality assumption
import scipy.stats as stats

residuals1 = (y1 - preds1)
sm.graphics.qqplot(residuals1, dist=stats.norm, line='45', fit=True);
```

It looks like our model's errors are generally homoscedastic.

```python
#checks homoscedasticity assumption
fig, ax = plt.subplots()

ax.scatter(preds1, residuals1, alpha=0.5)
ax.plot(preds1, [0 for i in range(len(x1))])
ax.set_xlabel("Predicted Value")
ax.set_ylabel("Actual - Predicted Value");
```

### Final Model

For our final model, we did a multiple linear regression model using living space square footage and dummy variables for number of bedrooms, condition of the house, year of the home sale, and average zipcode income bins.

Our model suggests that each square foot of living space is associated with on average about a \\$320 increase in home price all else equal.

A 3-bedroom home costs on average about \\$15,000 more than a 2-bedroom home all else equal.

Homes in good condition cost on average about \\$100,000 more than homes in bad condition all else equal.

Homes sold in 2022 cost on average about \\$110,000 more than homes sold in 2021.

Homes in zipcodes with average incomes of \\$100k-\\$150k cost on average about \\$180,000 more than homes in zipcodes with average incomes of \\$50k-\\$100k. And homes in zipcodes with average incomes of \\$150k-\\$200k cost on average about \\$85,000 more than homes in zipcodes with average incomes of \\$100k-\\$150k.

#### Rationale for Including Each Variable

Square footage of living space was the variable most correlated with price.

It's important to know how much to expect to pay for a 2-bedroom versus 3-bedroom home when calculating budget estimates.

Since this is an immediate relief program, we want homes that are in good enough condition to be bought and rented out in short order. But in the medium-term, the SSHA may want to run the numbers to see if buying a home in worse condition and doing some repairs is cheaper than buying a home in good condition.

Homes were selling for more in the first half of 2022 than in 2021. However, mortgage rates have since risen dramatically and home prices have begun to fall. We are assuming in our budget estimates that 2023 prices may be more similar to 2021 than 2022.

Homes tend to be more expensive in wealthier neighborhoods. This finding probably is related to wealthier neighborhoods having better schools, public services, and lower crime. It is a significant predictor of price. The SSHA can look at the different neighborhood price tiers and decide the best balance between maximizing quality of neighborhood and minimizing budget.

#### Model Interpretation

Our model had an R-squared of .473, meaning 47.3% of the variability in price can be explained by changes in our explanatory variables. Meanwhile, the RMSE for this model was about \\$215,000. Both figures are not remarkable by any means, but this is a marked improvement from our simple linear regression model.

```python
#creates final multiple linear regression model
y_final = df_final['price']
x_final = pd.concat([df_final['sqft_living'],df_final.iloc[:,-9:-7],df_final.iloc[:,-4:-1]],axis=1)
model_final = sm.OLS(y_final,sm.add_constant(x_final)).fit()
model_final.summary()
```

```python
#generates RMSE for final model
from statsmodels.tools.eval_measures import rmse
ypred = model_final.predict(sm.add_constant(x_final))
rmse_final = rmse(y_final, ypred)
rmse_final
```

Our final model generally passes the linearity assumption. It's not perfect but it appears to have improved from the simple model.

```python
#checks linearity assumption
preds_final = model_final.predict(sm.add_constant(x_final))
fig, ax = plt.subplots()

perfect_line = np.arange(y_final.min(), y_final.max())
ax.plot(perfect_line, linestyle="--", color="orange", label="Perfect Fit")
ax.scatter(y_final, preds_final, alpha=0.5)
ax.set_xlabel("Actual Price")
ax.set_ylabel("Predicted Price")
ax.legend();
```

Our final model violates the normality assumption, moreso than the previous model. We should take a look at the tails of the dataset.

```python
#checks normality assumption
residuals_final = (y_final - preds_final)
sm.graphics.qqplot(residuals_final, dist=stats.norm, line='45', fit=True);
```

Our final model has two variables with VIFs over 10, indicating we've violated the multicollinearity assumption. The two variables are living space square footage and the condition dummy variable.

```python
#checks multicollinearity assumption
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = [variance_inflation_factor(x_final.values, i) for i in range(x_final.shape[1])]
pd.Series(vif, index=x_final.columns, name="Variance Inflation Factor")
```

Our final model generally appears to have errors that are homoscedastic.

```python
#checks homoscedasticity assumption
fig, ax = plt.subplots()

ax.scatter(preds_final, residuals_final, alpha=0.5)
ax.plot(preds_final, [0 for i in range(len(x_final))])
ax.set_xlabel("Predicted Value")
ax.set_ylabel("Actual - Predicted Value");
```

## Regression Results

Based on our model, we can derive budget estimates for the Seattle Social Housing Authority. 

We calculated estimates for average-sized 2-3 bedroom homes in good condition for each income zipcode, ranging from \\$650 million at the low end to \\$930 million at the high end. 

We can also calculate rough rental market rates based on these home prices, which range from \\$2600-\\$3700 per month. Recall that our middle-income households can afford \\$2400-\\$3600, so we should be able to rent out these homes without discounting significantly.

Ultimately, the long-term fix to a limited affordable housing supply is to build more homes.

However, given the time it takes to identify land, acquire zoning permits, design and build homes, and the urgency of the crisis, buying 1000 single-family rental properties for middle-income families will provide some quick relief in the meantime.

This plan frees up entry-level housing for low-income households. And it will be the most cost-effective for Seattle, as middle-income households can afford the market rental rate for median-priced homes.
